#### Establishing Independent And Dependent Variables
* Independent : X = sepal_length, sepal_width, petal_length, petal_width
* Dependent : Y = species

```python
# Select the first four columns of the dataset that describe the features of the flower
X = df.iloc[:,:4].values
# Select the last column of the dataset 
Y = df['species'].values
# displaying first 10 entries of X and Y
print(f"X \n {X[:10]}")
print("-"*25)
print(f"Y \n {Y[:10]}")
```

#### Splitting Dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
# display
print(f"X_train \n{X_train}")
print("-"*25)
print(f"X_test \n{X_test}")
print("-"*25)
print(f"Y_train \n{Y_train}")
print("-"*25)
print(f"Y_test \n{Y_test}")
```

#### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

#### Building and Training the Naive Bayes Model on Training Set

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, Y_train)
```

#### Predicting Test Results

```python
Y_pred = classifier.predict(X_test)
print(f"Y_pred \n {Y_pred}")
```

#### Confusion Matrix and Accuracy

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
cm = confusion_matrix(Y_test, Y_pred)
print(f"Accuracy : {accuracy_score(Y_test, Y_pred)}")
print(f"Confusion Matrix:\n {cm}")
```
